```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import os
```

```python
os.getcwd()
```

```python
data=pd.read_csv('data.csv', nrows=100)
data.head()
```

```python
data.describe()
```

```python
y=data['Profit']
x1=data[['R&D Spend','Marketing Spend']] 
x1.head()
```

```python
x=sm.add_constant(x1)
```

```python
result=sm.OLS(y,x).fit()
result.summary()
```
